##### Column 추가하는 방법

```python
order['VIP']=order['amount']>=500
order
```

##### True, False 아닌 값을 column 추가 하는 방법

```python
order.loc[order['amount']>=500, 'level']='VIP'
order.loc[order['amount']<500, 'level']='Non-VIP'
order
```

#### apply: 판다스로는 못하겠는데, 파이썬으로 할 수 있을 때

```python
def is_vip(amount):
    if amount>=500:
        return 'VIP'
    else:
        return 'Non-VIP'

order['level']=order['amount'].apply(is_vip)
order
```
