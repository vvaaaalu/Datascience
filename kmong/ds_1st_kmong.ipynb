#### osversion 컬럼 분리

```python
#ostype 은 in으로 iOS가 있으면 iOS로, Android가 있으면 Android로 출력하도록한다.
#osversion은 S|d 를 기준으로 split해서 뒷부분만 출력하도록 한다. by me
```

```python
pd.DataFrame(raw_log['osversion']).head(2)
```

```python
def is_iOS(osversion):
    if 'iOS' in osversion:
        return 'iOS'
    else:
        return 'Android'
    
def find_osversion(osversion):
    return osversion.replace('iOS', '').replace('Android', '')

raw_log['ostype(clean)']=raw_log['osversion'].apply(is_iOS)
raw_log['osversion(clean)']=raw_log['osversion'].apply(find_osversion)

raw_log[['osversion','ostype(clean)', 'osversion(clean)']].head(5)
```

```python
def clean_devicemanufacturer(devicemanufacturer):
    if devicemanufacturer == 'samsung':
        return 'Samsung'
    elif devicemanufacturer == 'LGE':
        return 'LG'
    elif devicemanufacturer == 'LG Electronics':
        return 'LG'
    elif devicemanufacturer == 'Apple':
        return 'Apple'
    else:
        return 'Other'

raw_log['devicemanufacturer(clean)']=raw_log['devicemanufacturer'].apply(clean_devicemanufacturer)
raw_log['devicemanufacturer(clean)'].value_counts()
```

```python
def clean_channel(channel):
    return pd.isnull(channel)
   
raw_log['channel'].apply(clean_channel)
```

```python
def clean_channel(channel):
    if pd.isnull(channel):
        return channel
    if 'google' in channel:
        return 'google'
    elif 'daum' in channel:
        return 'daum'
    elif 'naver' in channel:
        return 'naver'
    elif 'apple' in channel:
        return 'apple'
    elif 'WEB' in channel:
        return 'web'
    else:
        return channel

raw_log['channel(clean)'] = raw_log['channel'].apply(clean_channel)
raw_log['channel(clean)'].value_counts()
```
